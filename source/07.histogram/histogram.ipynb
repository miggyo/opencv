```python
def show_histogram(img, title, xmax, ymax =None):
    color = ('b','g','r')
    for i, clr in enumerate(color):
        histr = cv2.calcHist([img],channels = [i],
                             mask = None, histSize = [256],ranges = [0,256])
        
        plt.plot(histr, color = clr)
        plt.xlim([0,xmax])
        if ymax != None : 
            plt.ylim([0,ymax])

    plt.title(title)
```

```python
show_histogram(bricks, 'HISTOGRAM FOR BLUE BRICKS', 256)
```

```python
show_histogram(dark_horse, 'HISTOGRAM FOR BLUE BRICKS', 256)
```

```python
show_histogram(dark_horse, 'HISTOGRAM FOR BLUE BRICKS', 15, 500000)
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize = (8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
lion_king = cv2.imread('../../data/lion_king.jpg',cv2.IMREAD_GRAYSCALE)
display_img(lion_king, cmap = 'gray')
```

```python
src = cv2.imread('../../data/field.bmp')
```

```python
src_ycrcb = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)
ycrcb_planes = cv2.split(src_ycrcb)
```

```python
print(type(ycrcb_planes))
```

```python
ycrcb_planes = np.array(ycrcb_planes)
```

```python
ycrcb_planes[0] = cv2.equalizeHist(ycrcb_planes[0])
dst_ycrcb = cv2.merge(ycrcb_planes)
dst = cv2.cvtColor(dst_ycrcb, cv2.COLOR_YCrCb2BGR)
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
